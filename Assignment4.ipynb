```python
%matplotlib notebook

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# read csv files 
df_al = pd.read_csv('./NHS_Canada/Alberta_2001_2011.CSV', 
                      skiprows=6, skipfooter=27)
df_bc = pd.read_csv('./NHS_Canada/BC_2001_2011.CSV', 
                      skiprows=6, skipfooter=27)
df_qb = pd.read_csv('./NHS_Canada/Quebec_2001_2011.CSV', 
                      skiprows=6, skipfooter=27)
df_on = pd.read_csv('./NHS_Canada/Ontario_2001_2011.CSV', 
                      skiprows=6, skipfooter=27)
df_2011 = pd.read_csv('./NHS_Canada/10541120180521033141_2001_to_2011.CSV', 
                      skiprows=6, skipfooter=27)

column_names = {'Place of birth (236)':'origin', 'Total - Citizenship [3]':'total', 'Total - Citizenship [4]':'total'}
df_al = df_al.iloc[:,0:2].rename(columns=column_names)
df_bc = df_bc.iloc[:,0:2].rename(columns=column_names)
df_qb = df_qb.iloc[:,0:2].rename(columns=column_names)
df_on = df_on.iloc[:,0:2].rename(columns=column_names)
df_2011 = df_2011.iloc[:,0:2].rename(columns=column_names)

def clean_df(df_2011, n_top):
    df_2011.sort_values('total', inplace=True, ascending=False)

    # split dataframe to countries and continents
    continents = ['Total - Place of birth of respondent',
                  '  Born in Canada',
                  '  Born outside Canada',
                  '    Americas',
                  '      North America',
                  '      Central America',
                  '      Caribbean and Bermuda',
                  '      South America',
                  '      Americas, n.i.e. [7]',
                  '    Europe',
                  '      Western Europe',
                  '      Eastern Europe',
                  '      Northern Europe',
                  '      Southern Europe',
                  '      Europe, n.i.e. [12]',
                  '    Africa',
                  '      Western Africa',
                  '      Eastern Africa',
                  '      Northern Africa',
                  '      Central Africa',
                  '      Southern Africa',
                  '      Africa, n.i.e. [15]',
                  '    Asia',
                  '      West Central Asia and the Middle East',
                  '      Eastern Asia',
                  '      Southeast Asia',
                  '      Southern Asia',
                  '      Asia, n.i.e. [24]',
                  '    Oceania',
                  '    Other']

    is_continent = [x in continents for x in df_2011['origin'].values]
    df_2011['is_continent'] = pd.Series(is_continent, index=df_2011.index.tolist())
    df_2011['origin'] = df_2011['origin'].apply(lambda x: x.strip())
    df_2011.set_index('origin', inplace=True)
    df_countries = pd.DataFrame(df_2011.loc[df_2011['is_continent']==False,'total'])

    row_names = {'United Kingdom [10]': 'United Kingdom',
                'China [20]': 'China',
                'Hong Kong Special Administrative Region': 'Hong Kong',
                'Iran [17]': 'Iran',
                'Korea, South [22]': 'Korea, South'}
    df_countries.rename(row_names, axis='index', inplace=True)
    #print(df_countries.head(10))
    #df_countries.plot.bar()
    df_top = df_countries.iloc[:n_top,:]
    #print(df_top)
    
    return df_top

df_altop = clean_df(df_al, 3)
df_bctop = clean_df(df_bc, 3)
df_qbtop = clean_df(df_qb, 3)
df_ontop = clean_df(df_on, 3)

def plot_df(df_top, color, province):
    rect = plt.bar(df_top.index.tolist(), height=df_top['total'], color=color)
    plt.yticks([])
    plt.xlabel('Countries', fontsize=10)
    plt.title('Immigration in {} between 2000 and 2011: \n Top three countries of origin'.format(province), fontsize=10)
    plt.ylim([0,155000])
    if province=='Quebec':
        [t.set_color('red') for t in plt.gca().xaxis.get_ticklabels()]

    for r in rect:
        height = r.get_height()
        plt.gca().text(r.get_x() + r.get_width()/2, height, '{}'.format(int(height)), ha='center', va='bottom')

    return

plt.figure(figsize=(10,8))
plt.subplot(2,2,1)
plot_df(df_altop, 'red', 'Alberta')

plt.subplot(2,2,2)
plot_df(df_bctop, 'blue', 'British Columbia')

plt.subplot(2,2,3)
plot_df(df_qbtop, 'green', 'Quebec')

plt.subplot(2,2,4)
plot_df(df_ontop, 'magenta', 'Ontario')

plt.tight_layout()
plt.show()
```

1. The regions of intererst are the provinces Alberta, British Columbia, Quebec and Ontario in Canada. While the category is Immigration.

2. Research Question: What are the three most common countries of origin of immigrants in these 4 provinces, and are these the same in all the four provinces?

3. The visual shows the top 3 countries of origin in the four provinces. We can see that for Alberta, British Columbia and Ontario, the top 3 countries of origin are Philippines, India and China. However, for Quebec the top 3 countries of origin are different from the other three provinces and they are Algeria, Morocco and France. The difference in the top 3 countries of immigrants in Quebec can be linked to the fact that it is a French speaking province and the top 3 countries of origin are Francophone as well.

1. I ensured the visual is truthful by making the y-axis uniform this ensures the relative heights of the graphs reflects the actual number of immigrants.
2. I ensured beauty in my visual by using fonts of a reasonable size and proportion and ensuring all elements of the visual are adequately spaced and there is no overlap.
3. I achieved functionality by using subplots to show the different provinces, as this would capture the information from the different provinces in one visual while ensuring each chart is not too busy.
4. I made the visual insightful by using a different color for the top 3 countries of origin in Quebec, since they are different from that of the other three provinces.
