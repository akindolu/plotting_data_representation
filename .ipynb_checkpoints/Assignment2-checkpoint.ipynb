```python
import matplotlib.pyplot as plt
import mplleaflet
import pandas as pd

def plot_temps(binsize, hashid):
    stations_filename, data_filename = download_files(binsize, hashid)
    
    #df = pd.read_csv('data/C2A2_data//BinnedCsvs_d{}/{}.csv'.format(binsize, hashid))
    df2 = pd.read_csv(data_filename)

    #print(df2.head())
    df2['Data_Value'] = df2['Data_Value'].apply(lambda x: x/10.0)
    df2['Date'] = pd.to_datetime(df2['Date'])
    df2.sort_values('Date', inplace=True)
    #print(df2.head(10))
    df3 = df2.groupby(['Date', 'Element'], as_index=False)[['Data_Value']].mean()
    #print(df3.shape)
    
    
    #print(df3.shape)
    #df3.dropna(axis=0, how='all', inplace=True)
    #df3.set_index(['Year', 'Month', 'Day'], inplace=True)
    #print(df3.shape)
    
    #print(df3.shape)
    df4 = df3.pivot(index = 'Date', columns='Element', values='Data_Value')
    df4.reset_index(inplace=True)
    #print(df4.shape)
    
    #Apply moving average to TMIN and TMAX
    #df4['TMIN'] = pd.rolling_mean(df4['TMIN'], window=1)
    #df4['TMAx'] = pd.rolling_mean(df4['TMAX'], window=1)
    
    df4['Year'] = df4['Date'].dt.year
    df4['Month'] = df4['Date'].dt.month
    df4['Day'] = df4['Date'].dt.day

    #remove leap year
    df4 = df4[~((df4.Month == 2) & (df4.Day == 29))]
    #print(df3.shape)

    #print(df4.head())
    df_05_14 = df4[df4['Date'] < '2015-01-01']
    df_15 = df4[df4['Date'] >= '2015-01-01']
    #print(df_05_14.head())
    #print(df_15.head())
    
    # daily max and daily min over 10 years (2005 t0 2014)
    daily_max = df_05_14.groupby(['Month', 'Day'])[['TMAX']].max()
    daily_min = df_05_14.groupby(['Month', 'Day'])[['TMIN']].min()
    #print('daily minimum temperatures: \n', daily_min.head())
    #print('daily maximum temperatures: \n', daily_max.head())
    
    # create new column for month and day
    daily_max.reset_index(inplace=True)
    daily_max['Year'] = 2015
    daily_max['Date'] = daily_max[['Year', 'Month', 'Day']].apply(lambda x: '{}-{}-{}'.format(x[0],x[1],x[2]), axis=1)
    daily_max['Date'] = pd.to_datetime(daily_max['Date'], format='%Y-%m-%d')
    daily_max.set_index(['Month', 'Day'], inplace=True)

    daily_min.reset_index(inplace=True)
    daily_min['Year'] = 2015
    daily_min['Date'] = daily_min[['Year', 'Month', 'Day']].apply(lambda x: '{}-{}-{}'.format(x[0],x[1],x[2]), axis=1)
    daily_min['Date'] = pd.to_datetime(daily_min['Date'], format='%Y-%m-%d')
    daily_min.set_index(['Month', 'Day'], inplace=True)
    #print(daily_max.head())
    
    # 2015 record breakers
    df_15_records = df_15.set_index(['Month', 'Day'])
    df_15_records['Tlower'] = df_15_records['TMIN'] < daily_min['TMIN']
    df_15_records['Thigher'] = df_15_records['TMAX'] > daily_max['TMAX']
    df_15_records = df_15_records[((df_15_records.Tlower == True) | (df_15_records.Thigher == True))]
    #print(df_15[((df_15.Tlower == True) | (df_15.Thigher == True))])
    
    x1 = df_05_14.set_index('Date').index
    y1_min = df_05_14['TMIN']
    y1_max = df_05_14['TMAX']
    
    x1_min = daily_min.set_index('Date').index
    x1_max = daily_max.set_index('Date').index
    y1_min = daily_min['TMIN']
    y1_max = daily_max['TMAX']

    x2_min = df_15_records[df_15_records['Tlower'] == True].set_index('Date').index
    x2_max = df_15_records[df_15_records['Thigher'] == True].set_index('Date').index
    y2_min = df_15_records[df_15_records['Tlower'] == True]['TMIN']
    y2_max = df_15_records[df_15_records['Thigher'] == True]['TMAX']
    
    plt.figure(figsize=(20,10))
    #plt.plot(df_05_14[df_05_14['Year'] == 2005]['Date'], df_05_14[df_05_14['Year'] == 2005]['TMAX'])
    #plt.plot(df_05_14[df_05_14['Year'] == 2005]['Date'], df_05_14[df_05_14['Year'] == 2005]['TMIN'])
    plt.plot(x1_min, y1_min, 'y-', x1_max, y1_max, 'r-')
    #plt.fill_between(df_05_14['Date'], df_05_14['TMIN'], df_05_14['TMAX'])
    plt.fill_between(x1_min, y1_min, y1_max, facecolor='grey')
    #plt.plot(df_15_low['Date'], df_15_low['TMIN'], 'ro', df_15_high['Date'], df_15_high['TMAX'], 'bo')
    plt.plot(x2_min, y2_min, 'ko', x2_max, y2_max, 'k*', markersize=10)
    plt.legend(['2005 to 2014 lowest daily temperature', '2005 to 2014 highest daily temperature', 
                '2015 temperature lower than decade lowest', '2015 temperature greater than decade highest'], 
              fontsize=15)
    plt.xlabel('Months of the year', fontsize=15)
    plt.ylabel('Temperature ($^\circ$C)', fontsize=15)
    plt.title('Maximum and minimum daily temperatures', fontsize=15)
    #plt.show()
    plt.savefig('Weather')
    
    return

plot_temps(400,'fb441e62df2d58994928907a91895ec62c2c42e6cd075c2700843b89')
```

```python
import matplotlib.pyplot as plt
import mplleaflet
import pandas as pd

def leaflet_plot_stations(binsize, hashid):
    stations_filename, data_filename = download_files(binsize, hashid)

    #df = pd.read_csv('data/C2A2_data/BinSize_d{}.csv'.format(binsize))
    #df = pd.read_csv('data/C2A2_data//BinnedCsvs_d{}/{}.csv'.format(binsize, hashid))
    df = pd.read_csv(stations_filename)

    station_locations_by_hash = df[df['hash'] == hashid]

    lons = station_locations_by_hash['LONGITUDE'].tolist()
    lats = station_locations_by_hash['LATITUDE'].tolist()
    
    plt.figure(figsize=(8,8))

    plt.scatter(lons, lats, c='r', alpha=0.7, s=200)
    
    return mplleaflet.display()

leaflet_plot_stations(400,'fb441e62df2d58994928907a91895ec62c2c42e6cd075c2700843b89')
```
